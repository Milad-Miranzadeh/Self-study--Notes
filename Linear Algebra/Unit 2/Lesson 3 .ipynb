## Span and linear independence examples
Let's say we have a set of vectors as below:
$$ S= \begin{bmatrix} 1 \\ -1 \\ 2 \end{bmatrix} \text{, }\begin{bmatrix} 2 \\ 1 \\ 3 \end{bmatrix} \text{, } \begin{bmatrix} -1 \\ 0 \\ 2 \end{bmatrix}$$
1. Doest the span of S is equall $\mathbb R^3$?
2. Are these vectos linearly independant?

## Q1.
If the span is eauall to $\mathbb R^3$, it means that some linear combination of these vectors should be able to construct any vector in $\mathbb R^3$. We sould be able to use some arbitrary constants and combine them in order to represent any vector in $\mathbb R^3$ So;
$$ S= C_1\begin{bmatrix} 1 \\ -1 \\ 2 \end{bmatrix} + C_2\begin{bmatrix} 2 \\ 1 \\ 3 \end{bmatrix} + C_3\begin{bmatrix} -1 \\ 0 \\ 2 \end{bmatrix}=\begin{bmatrix} a \\ b \\ c \end{bmatrix} $$
$$\rightarrow \begin{bmatrix} C_1+2C_2-C_3 \\ -C_1+C_2+0 \\ 2C_1+3C_2+2C_3 \end{bmatrix}=\begin{bmatrix} a \\ b \\ c \end{bmatrix}$$
$$E_1\rightarrow C_1+2C_2-C_3 =a $$ $$E_2\rightarrow-C_1+C_2=b $$ $$E_3\rightarrow2C_1+3C_2+2C_3=c $$
To solve this equation, we use the elimination method,so we add $E_1$ and $E_2$:
$$ 3C_2-C_3 = a+b$$
Then, we add the first and the third equations; the first equation is doubled so we have $2E_1$ and $E_3$:
$$-C_2 + 4C_3= c-2a $$
The finla product is $E_4$ and $E_5$
$$ E_4 \rightarrow C_1+2C_2-C_3=a $$ $$E_5\rightarrow 3C_2 -C_3 = a+b$$
We add $E_4$ and $3E_5$ to eliminate the $C_2$:
$$E_6 \rightarrow 11C_3 = b+c-5a \rightarrow C_3 = \frac{c+b-5a}{11}$$
$$ 33C_2 + (c+b-5a)= 11b + 11a \rightarrow C_2 = \frac{-c+10b+16a}{33}$$
$$ 33C_1 + 2C_2 + 3C_3 = 11a \rightarrow C_1= \frac{c-21b-16a}{33}$$ 
Based on the calculation, you can have any vector in $\mathbb R^3$ that there will be a combination of these constants that will add up to that vevctor. So:
$$\text{Span(S)} = \mathbb R^3 $$ 

## Q2. 
If the set is linearly independant, the only solution is below:
$$C_1\begin{bmatrix} 1 \\ -1 \\ 2 \end{bmatrix} + C_2\begin{bmatrix} 2 \\ 1 \\ 3 \end{bmatrix} + C_3\begin{bmatrix} -1 \\ 0 \\ 2 \end{bmatrix}=\begin{bmatrix} 0 \\ 0 \\ 0 \end{bmatrix} $$
and one of the constant must be not zero. This is the same as putting a, b, and c se zeros in the Q1. So; 
$$a=b=c=0 \rightarrow C_1= \frac{c-21b-16a}{33}\rightarrow C_1=0$$
$$a=b=c=0 \rightarrow C_2 = \frac{-c+10b+16a}{33} \rightarrow C_2=0$$
$$a=b=c=0 \rightarrow C_3 = \frac{c+b-5a}{11} \rightarrow C_3=0$$
In conclusion, the only way to have the zero matrix is that all the constant must be equall to zero, so this set of vectors is linearly independant.
