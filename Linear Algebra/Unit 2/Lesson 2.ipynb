# Introduction to linear independence
Let's say we have a set of vectors as below. We can calculate the sapan of these two vectors.
$${\overrightarrow a= \begin{bmatrix} 2 \\ 3 \end{bmatrix} \text{, } \overrightarrow b=\begin{bmatrix} 4 \\ 6 \end{bmatrix}}$$
$$\text{span }(\overrightarrow a, \overrightarrow b)= C_1\begin{bmatrix} 2 \\ 3 \end{bmatrix} +C_2 \begin{bmatrix} 4 \\ 6 \end{bmatrix} $$
Now $\overrightarrow b$ is 2 times of the $\overrightarrow a $ thus, we can say:
$$ C_1\begin{bmatrix} 2 \\ 3 \end{bmatrix} +2C_2 \begin{bmatrix} 2 \\ 3 \end{bmatrix}$$
$$ (C_1+2C_2)\begin{bmatrix} 2 \\ 3 \end{bmatrix} \rightarrow C_3 \begin{bmatrix} 2 \\ 3 \end{bmatrix} \text{, } C_3= C_1+2C_2$$
Even though we started we 2 vectors, the linear combination of these 2 vectors can be reduced to the scalar combination of 1 vector. All the linear combination of these vectors are going to lie on the line A. The span of the these vectors  is equal to line A;
$$ \text{span}(\begin{bmatrix} 2 \\ 3 \end{bmatrix} \text{, } \begin{bmatrix} 4 \\ 6 \end{bmatrix})= A$$
These two vectors are **co-linear**. The span of these co-linear vectors is reduced to one line and you can not represent of any other vector which is not in the mentioned line(c). This set is **linearly dependent set**. This means that one vector in that set($\overrightarrow b$) can be represent as a combination of the other vector($\overrightarrow a$) in the set and it will not add any new information. 

```python
import matplotlib.pyplot as plt
import numpy as np
a= np.array([2,3])
c=np.array([2,-2])
fig, ax = plt.subplots(figsize=(5,5))
ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
ax.grid(True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.quiver(0, 0, a[0], a[1], angles= 'xy', scale_units ='xy', scale=1, width=0.01, color= 'red', label= 'C3a')
ax.quiver(0, 0, c[0], c[1], angles= 'xy', scale_units ='xy', scale=1, width=0.01, color= 'green', label= 'b')
ax.axhline(y=0, color='k', linewidth=2)
ax.axvline(x=0, color='k', linewidth=2)
ax.axline((2,3), (4,6), color='yellow', linewidth=2, label='Span')
ax.legend()
plt.show()
```

In this example, the 2 vectors are not co-linear because they define 2 different dimensional in the $\mathbb R^2$. In the example below, we have 3 different vectors($V_1, V_2, V_3$). These vectors are  linearly dependent. This is because the sum of $V_1 \text{ and } V_2$ is equall to $V_3$. 
$$ \overrightarrow V_1= \begin{bmatrix} 2 \\ 3 \end{bmatrix} \text{, } \overrightarrow V_2= \begin{bmatrix} 7 \\ 2 \end{bmatrix} \text{, } \overrightarrow V_3=\begin{bmatrix} 9 \\ 5 \end{bmatrix}$$
$$\overrightarrow V_1 +\overrightarrow V_2=\overrightarrow V_3$$ 
Since these vectors are linearly dependent, one of the vectors is redundant.  
Span($V_1,V_2,V_3$)= $\mathbb R^2$


## More example of linear independence
Let's say we have a set of vectors as below. This set is linearly dependennt if and only if this equation is satisfy. 
$$ S= [V_1, V_2,V_3,...,V_n]$$
$$\text{linearly dependent} \leftrightarrow C_1V_1+C_2V_2+C_3V_3+....+C_nV_V = \begin{bmatrix} 0 \\.\\. \\ . \\ 0_n \end{bmatrix} \text{for some } C_i\not= 0 \text{ and at least one } C_i\text{ is not 0}$$
Soppuse we have  the set(S) as below and the equation betwen the $\overrightarrow V_1$ is like below;
$$ S= [V_1,V_2,V_3,...,V_n]$$
$$ \overrightarrow V_1= a_1.\overrightarrow V_1+a_2.\overrightarrow V_2+a_3.\overrightarrow V_3+...+a_n.\overrightarrow V_n$$
If we want to prove that this set of vector is linearly dependent;
$$ 0=-\overrightarrow V_1+  a_1.\overrightarrow V_1+a_2.\overrightarrow V_2+a_3.\overrightarrow V_3+...+a_n.\overrightarrow V_n$$
$$\rightarrow 0=a_1\overrightarrow V_1+  a_1.\overrightarrow V_1+a_2.\overrightarrow V_2+a_3.\overrightarrow V_3+...+a_n.\overrightarrow V_n \text{, } a_1=-1 \rightarrow TRUE$$
$$\text {if } C_1 \not= 0 \rightarrow V_1 + \frac{C_2}{c_1}V_2+\frac{C_3}{c_1}V_3+...+\frac{C_n}{c_1}V_n=0$$
$$ \rightarrow \frac{C_2}{c_1}V_2+\frac{C_3}{c_1}V_3+...+\frac{C_n}{c_1}V_n=- V_1 \rightarrow TRUE$$
So we can show the vector $V_1$ as a combination of other vectors. 

## Example 1
Let's say we have the set of vectors as below and we want to find whether they are linearly dependt or not.
$$ S= \begin{bmatrix} 2 \\ 1 \end{bmatrix} \text{, } \begin{bmatrix} 3 \\ 2 \end{bmatrix}$$ 
$$\rightarrow a_1V_1 + a_2V_2 = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$$
$$ \rightarrow a_1\begin{bmatrix} 2 \\ 1 \end{bmatrix} + a_2\begin{bmatrix} 3 \\ 2 \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix} $$
$$ a_1(2) + a_2(3)= 0 $$ 
$$a_1(1) + a_2(2) =0 $$
By calculating these 2 equation, we have:
$$ a_1 =0 \text{ and } a_2=0$$
Since at least one of the scalar can be 0, these two vectors are linearly independent.
