# Linear combination and span
All the linear combination of vectors are a linear combination. Let's say we have a number of vectors as below:
$$ V_1, V_2,V_3,...,V_n \in \mathbb R^2 $$
$$ c_1V_1+ c_2V_2 + c_3V_3+...+c_nV_n , [c_1,c_2,...,c_n \in \mathbb R^2]$$
## Example
If we have vector a and b as below, the linear combination of a and b is any constant times a plus any constant times b.
$$\overrightarrow a= \begin{bmatrix} 1 \\ 2 \end{bmatrix} \text{  and  }\overrightarrow b= \begin{bmatrix} 0 \\ 3 \end{bmatrix}$$
$$ 0. \overrightarrow a + 0.   \overrightarrow b = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$$
$$ 3. \overrightarrow a + -2.   \overrightarrow b = \begin{bmatrix} 3 \\ 0 \end{bmatrix}$$
The linear combination can be written as below:
$$C_1. \overrightarrow a + C_2.   \overrightarrow b = \begin{bmatrix} C_1  \\ 2C_1+ 3C_2 \end{bmatrix}$$
In order to find the set of all of the vectors by tacking linear combination. 

At first, we have to draw the vectors on the chart. Then, we scale up the vectors by the constant($C_1 \text{ and } C_2$). This is one of the linear combination of these vectors. You can represent any vector in $\mathbb R^2$ with the linear combination of these vectors.

```python
import matplotlib.pyplot as plt
import numpy as np
a= np.array([2,1])
b= np.array([0,3])
c= np.array([3,6])
d= np.array([0,-6])
fig, ax = plt.subplots(figsize=(8,8))
ax.set_xlim([-8,8])
ax.set_ylim([-8,8])
ax.grid(True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.quiver(0, 0, a[0], a[1], angles= 'xy', scale_units ='xy', scale=1, width=0.01, color= 'red', label= 'a')
ax.quiver(0, 0, b[0], b[1], angles= 'xy', scale_units ='xy', scale=1, width=0.01, color= 'yellow', label= 'b')
ax.quiver(0, 0, c[0], c[1], angles= 'xy', scale_units ='xy', scale=1, width=0.01, color= 'red', label= '3a')
ax.quiver(3, 6, d[0], d[1], angles= 'xy', scale_units ='xy', scale=1, width=0.01, color= 'yellow', label= '-2b')
ax.axhline(y=0, color='k', linewidth=2)
ax.axvline(x=0, color='k', linewidth=2)
#ax.axline((2,1), (0,0), color='yellow', linewidth=2, label='S')
#ax.axline((0,1), (-2,0), color='blue', linewidth=2, label='L')
ax.legend()
plt.show()
```

We can use any scalar with any arbitary vectors to find the linear combination and we can fill up any of these gaps in $\mathbb R^2$ with the combination of a and b. We can write it as the **span** of the vectors a and b. 
$$ \text{Span }(\overrightarrow a, \overrightarrow b) = \mathbb R^2 $$
The span of a set of vectors ($V_1,V_2,V_3,...,V_n$) is the set of all of the vectors that are scale up.
$$ \text{Span }(V_1,V_2,V_3,...,V_n)= [C_1V_1+C_2V_2+C_3V_3+....+C_nV_n | C_i \in \mathbb R^2 \text{ and }{0<= i <= n}] $$
## Example
If we have another vectors such as below, the span is the line A.
$$ \overrightarrow a = \begin{bmatrix} 2 \\ 2 \end{bmatrix} \text{ and } \overrightarrow b= \begin{bmatrix} -2 \\ -2 \end{bmatrix}$$
$$ \text{Span }(\overrightarrow a, \overrightarrow b) = A \text{, } A= x $$
In this example, these vectors can not be represented in $\mathbb R^2$. E.g. the sapn of the zero vector is $\begin{bmatrix} 0 \\ 0 \end{bmatrix}$.
The span of the $\hat i$ and $\hat j$ is calculated as below:
$$ \hat i = \begin{bmatrix} 1 \\ 0 \end{bmatrix} \text{ and } \hat j = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$$

```python
import matplotlib.pyplot as plt
import numpy as np
a= np.array([2,2])
b= np.array([-2,-2])
fig, ax = plt.subplots(figsize=(8,8))
ax.set_xlim([-8,8])
ax.set_ylim([-8,8])
ax.grid(True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.quiver(0, 0, a[0], a[1], angles= 'xy', scale_units ='xy', scale=1, width=0.01, color= 'red', label= 'a')
ax.quiver(0, 0, b[0], b[1], angles= 'xy', scale_units ='xy', scale=1, width=0.01, color= 'yellow', label= 'b')
ax.axhline(y=0, color='k', linewidth=2)
ax.axvline(x=0, color='k', linewidth=2)
ax.axline((2,2), (0,0), color='green', linewidth=2, label='A')
ax.legend()
plt.show()
```

If we have 2 vectors such as a and b, we can represent any point (x) with some combination of a and b.
$$ C_1 \overrightarrow a + C_2 \overrightarrow b= \overrightarrow X = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}$$
$$ C_1 \begin{bmatrix} 1 \\ 2 \end{bmatrix} + C_2 \begin{bmatrix} 0 \\ 3 \end{bmatrix} = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}$$
$$ C_1  = x_1$$ $$ 2C_1+ 3C_2 = x_2$$
$$ \rightarrow 3C_2 = x_2 -2x_1$$
$$ \rightarrow C_2 = \frac{1}{3}(x_2 -2x_1)\text{ and } C_1 = x_1 \text{, } (x_1,x_2) \in \mathbb R^2$$
If we want to know what combination of a and b we want to get to the $\begin{bmatrix} 2\\2 \end{bmatrix}$:
$$ x_1 = 2 \rightarrow C_1 =2$$
