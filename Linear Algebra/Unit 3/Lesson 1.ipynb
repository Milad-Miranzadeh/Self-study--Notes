## Linear subspaces
The linear subspace of $R^n$ is defined as if a set of vectors (V) is a subset of $R^n$, in order for V to be a linear subspace of $R^n$, these three criteria should be met:
1. V contains zero vector.
2. If $\overrightarrow x$ is in V, then $C_i  .\overrightarrow x \in V$.
3. If $\overrightarrow a$ and  $\overrightarrow b$ are in V, then $\overrightarrow a + \overrightarrow b \in V$
