# Unit vector (Intro)
Another notation for representing vectors is to use the unit vector. The unit vectors in a 2-dimension are defined as below; we show this unit vectors with $\hat i$ and $\hat j$.
$$ \hat i= \begin{bmatrix} 1 \\0 \end{bmatrix} \text { , } \hat j= \begin{bmatrix} 0\\ 1 \end{bmatrix}$$
Any 2-dimensional vector can be shown as sum of these unit vectors. If we have a  $\overrightarrow v$ with (2,3) altitude and we want to represent it with the unit vectors. We can say the $\overrightarrow V$ is the sum of a vector that purely moves 2 units in the horizontal direction and a vector that moves 3 units in the vertical direction, so we can scale up the unit vectors:
$$ \overrightarrow v= 2\hat i + 3\hat j = \begin{bmatrix} 2 \\ 3 \end{bmatrix}$$ 
We can do the vector operations with this type of notation. E.g. we have these 2 vectors as below:
$$ \overrightarrow a = 2\hat i+ 3\hat j \text{ , } \overrightarrow b= -1\hat i+ 4\hat j$$
$$ \overrightarrow a + \overrightarrow b = (2-1)\hat i + (3+4) \hat j \rightarrow \overrightarrow a + \overrightarrow b = \hat i + 7\hat j$$

# Parametric representation of lines
Let's define some vectors as below; all the vectors are belonged to the $R^2$. 
$$ \overrightarrow v = \begin{bmatrix} 2\\ 1\end{bmatrix}$$
The $\overrightarrow v$ in the standard position can be seen in the chart below. Next, we define as S:
$$ S= [c\overrightarrow v| \text{ c} \in \mathbb {R}]$$
The graphical representation of this set is as below. 
C can be any number if it is a real number, if we can plot all the possible vectors from the standard postion, we have a bunch of vectors, all lined up in the line c. So, the set S is a **Co-linear vectors**. If we  view these vectors as a representation of a point in space in the $ \mathbb{R}$.
We can also show this in different starting position($\overrightarrow x$), so another line that is parllel to the first line can be drawn(L).
The "L" line, can be shown as below:
$$ L= [\overrightarrow x + t\overrightarrow V | \text{ t}\in \mathbb{R}]$$
> **This approach is the same as finding the line with the mx+b equation, but in higher dimensions, we use this way to find the set line(s).**
>
In Algebra 1, we find the line that goes through these 2 vectors as below;
$$ \overrightarrow a = \begin{bmatrix} 2 \\ 1 \end{bmatrix} \text{ and } \overrightarrow b= \begin{bmatrix} 0 \\ 3 \end{bmatrix}$$
$$ m= \frac{3-1}{0-2}= -1 \text { and } b= y-mx \rightarrow b= 3-(-1).(0) \rightarrow b= 3 $$
$$ \rightarrow y=-x +3 $$

```python
import matplotlib.pyplot as plt
import numpy as np
v= np.array([2,1])
fig, ax = plt.subplots(figsize=(7,7))
ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
ax.grid(True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.quiver(0, 0, v[0], v[1], angles= 'xy', scale_units ='xy', scale=1, width=0.01, color= 'red', label= 'v')
ax.axhline(y=0, color='k', linewidth=2)
ax.axvline(x=0, color='k', linewidth=2)
ax.axline((2,1), (0,0), color='yellow', linewidth=2, label='S')
ax.axline((0,1), (-2,0), color='blue', linewidth=2, label='L')
ax.legend()
plt.show()
```

In the second approach, first we find the vectors in the graph. The vector $\overrightarrow b - \overrightarrow a$ can be seen in the graph. Now, there is a line with a **t** in the space that goes through the $\overrightarrow b - \overrightarrow a$ ,so the $t(\overrightarrow b - \overrightarrow a)$ is a line in the standard position which need to be shifted in order to go through the vector $\overrightarrow b - \overrightarrow a$. We can add the vector $\overrightarrow b$  or vecotr $\overrightarrow a$ and it will lay on the orange vector and we can find the line **L** that goes thorugh $\overrightarrow a \text{ and } \overrightarrow b$.
$$L= \overrightarrow b + t(\overrightarrow b - \overrightarrow a)|\text{ t} \in \mathbb {R}$$
$$ \rightarrow L= \begin{bmatrix} 0\\3 \end{bmatrix} + t\begin{bmatrix} -2\\2 \end{bmatrix}|\text{ t} \in \mathbb{R}$$
The co-ordinate of the line L is $\begin{bmatrix} L_1\\L_2\end{bmatrix}$:
$$ L_1 = 0+t(-2) \text { and } L_2= 3+t(2)$$

```python
import matplotlib.pyplot as plt
import numpy as np
a= np.array([2,1])
b= np.array([0,3])
c= np.array([-2,2])
d= np.array([-2,2])
e= np.array([0,3])
fig, ax = plt.subplots(figsize=(7,7))
ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
ax.grid(True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.quiver(0, 0, a[0], a[1], angles= 'xy', scale_units ='xy', scale=1, color= 'red', label= 'a')
ax.quiver(0, 0, b[0], b[1], angles= 'xy', scale_units ='xy', scale=1, color= 'blue', label= 'b')
ax.quiver(2,1 , c[0], c[1], angles= 'xy', scale_units ='xy', scale=1, width= 0.01, color= 'orange', label= 'b-a')
ax.quiver(0,0 , d[0], d[1], angles= 'xy', scale_units ='xy', scale=1, width= 0.01, color= 'orange')
ax.quiver(-2,2 , e[0], e[1], angles= 'xy', scale_units ='xy', scale=1, color= 'blue')
ax.axhline(y=0, color='k', linewidth=2)
ax.axvline(x=0, color='k', linewidth=2)
ax.axline((0,0),(-2,2), color='yellow',linewidth=2)
ax.axline((-2,5),(-4,7), color='green',linewidth=2, label='L')
ax.legend()
plt.show()
```

Overall, if we have 2 vectors($\overrightarrow a \text{ and } \overrightarrow b$), the line($\overrightarrow L $) that goes thorugh these two vectors can be calculated as below:
$$\overrightarrow a = \begin{bmatrix} x_1 \\ y_1 \end{bmatrix} \text{ and } \overrightarrow b= \begin{bmatrix} x_2\\y_2 \end{bmatrix} $$
$$\rightarrow \overrightarrow L= \begin{bmatrix} x_1\\y_1 \end{bmatrix}+ t\begin{bmatrix} x_2-x_2\\y_2-y_1 \end{bmatrix}$$\
$$\rightarrow \overrightarrow L= \begin{bmatrix} L_1\\L_2 \end{bmatrix}$$ $$L_1= x_1 + t(x_2-x_1)$$ $$L_2= y_1 + t(y_2-y_1) $$
> In this equatio, one of the vectors must be used to calculate the $\overrightarrow L$.

This calculation can be useful when we have 3 dimention vectors.
$$ P_1= \begin{bmatrix} -1\\2\\7 \end{bmatrix} \text{ and } P_2 =\begin{bmatrix} 0\\3\\4 \end{bmatrix} \text{ ,} P_1 \text{ and } P_2 \in \mathbb R^3$$
$$ \overrightarrow L= \begin{bmatrix} 0\\3\\4 \end{bmatrix}+ t \begin{bmatrix} 0-(-1)\\3-2\\4-7 \end{bmatrix}$$
$$ \overrightarrow L =\begin{bmatrix} L_1\\L_2\\L_3 \end{bmatrix}$$
$$L_1= 0 + t(0-(-1)) \rightarrow L_1= t$$
$$L_2= 3 + t(3-2) \rightarrow L_2= 3+t$$
$$L_3= 4 + t(4-7) \rightarrow L_3= 4-3t$$
$$ \rightarrow \overrightarrow L = \begin{bmatrix} t\\3+t\\4-3t \end{bmatrix} | \text{ t} \in \mathbb R^3$$
>** If you use the $(P_1 -P_2)$, you have to use $P_1$ for the next step, and if you use the $(P_2 -P_1)$, you have to use $P_2$ for the next step**
