# Vector Examples
These vectors are all two tuples and follow the $R^2$. The vectors provided in the example is represent as below:
$$ a= \begin{bmatrix} -1 \\ 2 \end{bmatrix} \text {, } b= \begin{bmatrix} 3 \\ 1 \end{bmatrix} \text{, } a+b = \begin{bmatrix} 2 \\ 3 \end{bmatrix}$$
For drawing a vector, you can start from any point you want. E.g. we can start from the $(x_1,x_2)$ so, the ending point for vector a is $(x_1 -1, x_2 + 2)$. If the starting point is (-4,4) then the ending point is (-5,6). The best starting position is the standard position.
> The **standard starting position** For drawing a vector is the (0,0).

The vectors b and a+b in the standard position is demonstrate as below. To understand how the a+b vector is determined, a valid approach is to put vector b at the end of the vector a. 
> Even though vector b is not in the standard position, the representation of vector a+b is valid!



```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator
a = np.array([-1, 2])
b = np.array([3, 1])
c = np.array([2, 3])
a_1= np.array([-1, 2])
b_1= np.array([3, 1])
c_1= np.array([2, 3])
fig, ax = plt.subplots(figsize=(7, 7))
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.yaxis.set_major_locator(MultipleLocator(1))
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.grid(True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.axhline(y=0, color='k', linewidth=2)
ax.axvline(x=0, color='k', linewidth=2)
ax.quiver(-4, 4, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='a')
ax.quiver(-4, 4, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='red', label='b')
ax.quiver(-4, 4, c[0], c[1], angles='xy', scale_units='xy', scale=1, color='green', label='a+b')
ax.quiver(0, 0, a_1[0], a_1[1], angles='xy', scale_units='xy', scale=1, color='blue', label= 'standard a')
ax.quiver(-1, 2, b_1[0], b_1[1], angles='xy', scale_units='xy', scale=1, color='red', label='standard b')
ax.quiver(0, 0, c_1[0], c_1[1], angles='xy', scale_units='xy', scale=1, color='green', label='standard a+b')
ax.legend()
plt.show()
```

In this section, we want to see waht happend when we multiply a scalar to a vector. Let's say we have vector E as (1,2). The vectors F & G are calculated as below;
$$ E= \begin{bmatrix} 1 \\ 2 \end{bmatrix} \text{ , } F=(2.E)= \begin{bmatrix} 2 \\ 4\\ \end{bmatrix} \text{, }  G= ({-4}.E)=\begin{bmatrix} -4 \\ -8 \\ \end{bmatrix}$$  $$F=\begin{bmatrix} 2 \\ 4 \\ \end{bmatrix} \text{, }H= (-\frac{1}{2}.F)= \begin{bmatrix} -1 \\ -2 \end{bmatrix} $$
As it mentioned before, we can draw vectors F & G in any point but the standard position is to draw it from (0,0). 
If we want to determine the differentiation between F and H, we can say;
$$ I= F-H \rightarrow F + (-1.H) \Rightarrow \begin{bmatrix} 2 \\ 4 \\ \end{bmatrix} + \begin{bmatrix} 1 \\ 2 \end{bmatrix} \rightarrow \begin{bmatrix} 3 \\ 6 \end{bmatrix}$$
> In order to have a crystal clear visualization of the vectors, the resulted vectors (F, G, H, and I) are transfered to different position so you can understand the effect and multiplying, fraction, and differentiation.


```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator
E= np.array([1, 2])
F= np.array([2, 4])
G= np.array([-4, -8])
H= np.array([-1,-2])
I= np.array([3,6])
fig, ax = plt.subplots(figsize=(7, 7))
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.yaxis.set_major_locator(MultipleLocator(1))
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.grid(True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.axhline(y=0, color='k', linewidth=2)
ax.axvline(x=0, color='k', linewidth=2)
ax.quiver(0, 0, E[0], E[1], angles='xy', scale_units='xy', scale=1, color='blue', label='E')
ax.quiver(1, 0, F[0], F[1], angles='xy', scale_units='xy', scale=1, color='red', label='F')
ax.quiver(0, 0, G[0], G[1], angles='xy', scale_units='xy', scale=1, color='green', label='G')
ax.quiver(-1, 0, H[0], H[1], angles='xy', scale_units='xy', scale=1, color='yellow', label='H')
ax.quiver(2, 0, I[0], I[1], angles='xy', scale_units='xy', scale=1, color='orange', label='I')
ax.legend()
plt.show()
```

If we want to visualise the differentiation; let's say we have the vectors A and B as below;
$$ A= \begin {bmatrix} 2 \\ 3 \end{bmatrix} \text { and } B= \begin{bmatrix} -4 \\ -2 \end{bmatrix}$$
The differentiation is calculated as below which is notated as vector c;
$$ A-B = \begin {bmatrix} 2 \\ 3 \end{bmatrix} -  \begin{bmatrix} -4 \\ -2 \end{bmatrix} \rightarrow C= \begin{bmatrix} 6 \\ 5 \end{bmatrix}$$
The differentiation between B and A (D=B-A) is calculated as below;
$$ B-A = \begin {bmatrix} -4 \\ -2 \end{bmatrix} -  \begin{bmatrix} 2 \\ 3 \end{bmatrix} \rightarrow D= \begin{bmatrix} -6 \\ -5 \end{bmatrix}$$
It is obvious that the vector D is pointing the opposite direction of the vector C

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator
A= np.array([2, 3])
B= np.array([-4, -2])
C= np.array([6, 5])
D= np.array([6,5])
K= np.array([-6,-5])
fig, ax = plt.subplots(figsize=(7, 7))
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.yaxis.set_major_locator(MultipleLocator(1))
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.grid(True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.axhline(y=0, color='k', linewidth=2)
ax.axvline(x=0, color='k', linewidth=2)
ax.quiver(0, 0, A[0], A[1], angles='xy', scale_units='xy', scale=1, color='blue', label='A')
ax.quiver(0, 0, B[0], B[1], angles='xy', scale_units='xy', scale=1, color='red', label='B')
ax.quiver(1, 0, C[0], C[1], angles='xy', scale_units='xy', scale=1, color='green', label='C')
ax.quiver(-4, -2, D[0], D[1], angles='xy', scale_units='xy', scale=1, color='yellow', label='A-B')
ax.quiver(1, 0, K[0], K[1], angles='xy', scale_units='xy', scale=1, color='ORANGE', label='B-A')
ax.legend()
plt.show()
```
