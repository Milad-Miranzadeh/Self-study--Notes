# Adding vectors algebraically & graphically 
## Adding vectors algebraically  
Suppose we have these 2 vectors a and b with the coordinates as below. How we can add these two vectors together?
$$ \overrightarrow a= \begin{bmatrix} 6 \\ -2 \end{bmatrix} , \overrightarrow b= \begin{bmatrix} -4 \\ 4 \end{bmatrix}, {\overrightarrow a,\overrightarrow b}\in R^2 $$
To add these vectors, we can add each coordinates together, so the magnitudes in the x-axis and the magnitudes in the y-axes will add to each other respectively.
$$ \overrightarrow a+ \overrightarrow b = \begin{bmatrix} 6-4 \\ -2+4 \end{bmatrix} = \begin{bmatrix} 2 \\ 2 \end{bmatrix}$$


## Adding vectors graphically
These two labels can be demonstrate as below. If we want to add these two vectors, we can put the b vector at the ending point of vector a. This will result in the $b_1$ vector which can be seen in the plot. Now, the final vector (a+b) is make sense because the sum of those 2 vectors is the result where you start with vector a and what you start with vector b at the tail of vector a. 
> **Based on the definition, a+b and b+a is equal.**



```python
import matplotlib.pyplot as plt
import numpy as np
a= np.array ([6,-2])
b= np.array ([-4,4])
c= np.array([2,2])
d= np.array([-4,4])
fig, ax = plt.subplots(figsize=(4,4))
ax.set_xlim([-5,7])
ax.set_ylim([-3,5])
ax.grid(True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.quiver(0, 0, a[0], a[1], angles= 'xy', scale_units ='xy', scale=1, color= 'red', label= 'a')
ax.quiver(0 ,0 , b[0], b[1], angles= 'xy', scale_units ='xy', scale=1, color= 'green', label= 'b')
ax.quiver(0 ,0 , c[0], c[1], angles= 'xy', scale_units ='xy', scale=1, color= 'blue', label= 'a+b')
ax.quiver(6 ,-2 , d[0], d[1], angles= 'xy', scale_units ='xy', scale=1, color= 'green', label= 'b1')
ax.axhline(y=0, color='k', linewidth=2)
ax.axvline(x=0, color='k', linewidth=2)
ax.legend()
plt.show()
```

# Multiplying a vector by a scalar
Let's say we have a vector with [2,1] Coordinate. We can show this vector as in the graph as below. If we want to multiply a vector by a scalar (3), we can write it as below:
$$ (\overrightarrow a.3)= 3.  \begin{bmatrix} 2 \\ 1 \end{bmatrix}$$
When we have this type of multiplying, the scalar is multiplying in each component of the vector so;
$$ \overrightarrow a .3 = \begin{bmatrix} 6 \\ 3 \end{bmatrix}$$
> When we multiply a vector by a scaler, it scales up the vector. It changes the magnitude but it doesn't change the direction. 

We can show this in the graph as below

```python
import matplotlib.pyplot as plt
import numpy as np
a= np.array([2,1])
b= np.array([6,3])
fig, ax= plt.subplots(figsize=(4,4))
ax.set_xlim([-1,7])
ax.set_ylim([-1,4])
ax.grid(True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.axhline(y=0, color='k', linewidth=2)
ax.axvline(x=0, color='k', linewidth=2)
ax.quiver(0,0, a[0],a[1], angles='xy', scale_units= 'xy', scale=1, color='blue', width=0.018, headwidth=2, headlength=4, label='a')
ax.quiver(0,0, b[0],b[1], angles='xy', scale_units= 'xy', scale=1, color='red', label='3a')
ax.legend()
plt.show()
```

If we want to multiply the vector with a negative scalar, (-1), the magnitude and the direction is changed. 
$$ \overrightarrow a .(-1) = \begin{bmatrix} -2 \\ -1 \end{bmatrix}$$

```python
import matplotlib.pyplot as plt
import numpy as np
a= np.array([2,1])
b= np.array([-2,-1])
fig, ax= plt.subplots(figsize=(4,4))
ax.set_xlim([-3,4])
ax.set_ylim([-2,2])
ax.grid(True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.axhline(y=0, color='k', linewidth=2)
ax.axvline(x=0, color='k', linewidth=2)
ax.quiver(0,0, a[0],a[1], angles='xy', scale_units= 'xy', scale=1, color='blue', label='a')
ax.quiver(0,0, b[0],b[1], angles='xy', scale_units= 'xy', scale=1, color='red', label='3a')
ax.legend()
plt.show()
```
