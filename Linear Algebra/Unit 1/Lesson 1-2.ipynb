# Vector for linear Algebra
A vector is something that has both magnitude and direction. If we have a car that goes with 5mph, it is not a vector because we don't have the direction. The speed by itself is not a vector quantity and it's a **scalar** quantity. If we want to turn it to a vector, we have to add the direction. For example, if we say the car is traveling 5 mph due east, it is now a vector quantity and it is called velocity.

One way to show this quantity is to draw an arrow with 5 units long, if we consider each unit represents 1 mph. We can consider the right side of the graph as the east side. So we can demonstrate this as below


```python
import matplotlib.pyplot as plt
import numpy as np

# Define a vector [2, 3]
v = np.array([5, 0])
v2= np.array([5,0])
v3= np.array([3,4])
# Create a figure and an axes object
fig, ax = plt.subplots(figsize=(4, 4))

# Set axis limits and grid
ax.set_xlim([-1, 6])
ax.set_ylim([-1, 6])
ax.grid(True)

# Add a title and axis labels
ax.set_xlabel('Direction')
ax.set_ylabel('Speed')
ax.set_title('Vectors')
ax.axhline(y=0, color='k', linewidth=2)
ax.axvline(x=0, color='k', linewidth=2)
# Draw the vector using quiver
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='V1')
ax.quiver(-1, 1, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label='v2')
ax.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color='green', label='v3')
# Add a legend
ax.legend()

plt.show()
```

When it comes to vectors, the starting point and the ending point are not important. So in the chart above, both V1 and V2 are the same vector even though the points are different. In the chart above, both V1 and V2 have the same magnitude and direction even though the starting points are totally different. The way we show vectors is to specify how much the arrow is moving in each directions.If we want to show $V_1$, we write it as below:
$$ V_1=(5,0)  = \begin{bmatrix} 5 \\ 0 \end{bmatrix}$$
The first coordinate shows the movement in the x-axis the second one shows the movement in the y-axis. If we want to show the $V_3$ in the mathematical notation, we can write it as below:
$$ V_3= \begin{bmatrix} 3 \\ 4 \end{bmatrix}$$

## Real coordinate spaces
Real coordinate space which is indicated as $R^2$ is an abstract of the representation of the 2-dimensional space. The real 2-dimensional is all the possible real valued 2 tuple. A **tuple** is an ordered list of values. So, the $ \begin{bmatrix} 3 \\ 4 \end{bmatrix}$ is 2-tuple. In the real 2-tuple, are the values must follow the real numbers : 
$$ V_1= \begin{bmatrix} 3 \\ 4 \end{bmatrix} \in R^2 $$
$$ V_1= \begin{bmatrix} i \\ 4 \end{bmatrix} \notin R^2 $$

> The $R^3$ is the all real 3 real tuples. E.g. $\begin{bmatrix} 3 \\ 0 \\ 4 \end{bmatrix}$
